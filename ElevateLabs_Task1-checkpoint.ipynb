```python
# 1. Convert 'Sex' to numerical
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# 2. Robust Title extraction from 'Name'
def extract_title(name):
    try:
        return name.split(',')[1].split('.')[0].strip()
    except:
        return 'Unknown'

df['Title'] = df['Name'].apply(extract_title)

# 3. Standardize the 'Title' column
title_mapping = {
    'Mr': 1, 'Miss': 2, 'Mrs': 3, 'Master': 4,
    'Dr': 5, 'Rev': 5, 'Col': 5, 'Major': 5,  
    'Unknown': 5  
}
df['Title'] = df['Title'].map(title_mapping).fillna(5)

# 4. Create family features
df['FamilySize'] = df['Siblings/Spouses Aboard'] + df['Parents/Children Aboard']
df['IsAlone'] = (df['FamilySize'] == 0).astype(int)

# 5. Feature scaling
scaler = StandardScaler()
numeric_cols = ['Age', 'Fare', 'Pclass', 'FamilySize', 'Title']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# 6. Outlier treatment (example for Fare)
q1, q3 = df['Fare'].quantile([0.25, 0.75])
iqr = q3 - q1
df['Fare'] = np.where(df['Fare'] > (q3 + 1.5*iqr), (q3 + 1.5*iqr), df['Fare'])

# Show final dataframe
print("\nFinal Features:")
print(df.head())
print("\nColumn Types:")
print(df.dtypes)
```

```python
# 1. First visualization BEFORE outlier treatment
plt.figure(figsize=(12,6))
sns.boxplot(data=df[['Age', 'Fare', 'FamilySize']])
plt.title("Feature Distributions Before Outlier Treatment")
plt.show()

# 2. Define numeric columns for outlier treatment
numeric_cols = ['Age', 'Fare', 'FamilySize']
```

```python
# 3. Remove outliers using IQR method
def remove_outliers(df, col):
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Create a copy before removing outliers to compare
df_clean = df.copy()
for col in numeric_cols:
    df_clean = remove_outliers(df_clean, col)
```

```python
# 4. Visualize AFTER outlier treatment
plt.figure(figsize=(12,6))
sns.boxplot(data=df_clean[['Age', 'Fare', 'FamilySize']])
plt.title("Feature Distributions After Outlier Removal")
plt.show()

# 5. Show how many rows were removed
print(f"Original dataset size: {len(df)}")
print(f"After outlier removal: {len(df_clean)}")
print(f"Rows removed: {len(df)-len(df_clean)}")

# 6. Correlation heatmap of cleaned data
plt.figure(figsize=(10,8))
numeric_df = df_clean.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title("Feature Correlations After Outlier Removal")
plt.tight_layout()
plt.show()
```
